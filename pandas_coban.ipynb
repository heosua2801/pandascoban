## Chọn cột bằng iloc

```python
# Truy xuất hàng dữ liệu theo chỉ mục trong DataFrame (Chỉ mục 3)
cauthu.iloc[3]
```

```python
# Truy xuất hàng dữ liệu theo chỉ mục trong DataFrame (Chỉ mục 0,1,5)
cauthu.iloc[[0,1,5]]
```

```python
# Truy xuất hàng dữ liệu theo chỉ mục trong DataFrame (Chỉ mục 5 đến 9 (10-1) )
cauthu.iloc[5:10]
```

```python
# Truy xuất dữ liệu ở hàng thứ (5), cột thứ (1) trong DataFrame
cauthu.iloc[5,1]
```

```python
# Lấy 5 hàng đầu tiên và 10 cột đầu tiên của DataFrame
cauthu.iloc[:5,:10]
```

## Chọn cột bằng loc

```python
# Truy xuất hàng dữ liệu tại chỉ mục 50 trong DataFrame
cauthu.loc[50]
```

## Index

```python
# Truy xuất thông tin về thứ hạng (Rk), tên cầu thủ (Player), và đội bóng (Squad) cho các hàng có chỉ mục 31, 10, và 2024
cauthu.loc[[31,10,2024],["Rk", "Player", "Squad"]]
```

```python
# Thiết lập cột Rk làm chỉ mục cho DataFrame
cauthu.set_index(["Rk"], inplace = True)
```
